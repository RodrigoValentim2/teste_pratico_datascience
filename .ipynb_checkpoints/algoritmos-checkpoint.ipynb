# Neste notebook é feito um experimento com os algoritmos  LinearSVC MultinomialNB e  LogisticRegression.

* Os experimentos foram feitos nos dados data1 que não consideram a culuna IdSentenca e no data2 que leva em consideração a coluna. Por fim, é feito um teste de Friedman pra vericar se os modelos obtêm resultados diferentes.




```python
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np
import warnings
from sklearn.model_selection import train_test_split

from sklearn.model_selection import KFold
from sklearn.model_selection import RepeatedKFold
import os

from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.svm import LinearSVC 

from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics as mt
from matplotlib import pyplot as plt


from numpy.random import seed
from numpy.random import randn
from scipy.stats import friedmanchisquare

%matplotlib inline
warnings.simplefilter("ignore", UserWarning)
```

```python
data = pd.read_csv("Outputs/data1.csv")
y = data.Sentimento.values
```

```python
sentimentos = data.groupby('Sentimento').Sentimento.count()
```

## Modelos

```python
models = [
    LinearSVC(),
    MultinomialNB(),
    LogisticRegression(random_state=0),
]
```

# Passos para o treinamento Treinamento:
* Bag of words model
* Dividiu-se o os dados utilizando validação cruzada com o K-fold, repetiu o processo duas vezes para que fosse possível fazer um teste de Friedman para comparar se os classifcadores tiveram resultados diferentes.

```python
def train(data):
    X = data.Texto.values
    count_vect = CountVectorizer()
    X_counts = count_vect.fit_transform(X)
    tfidf_transformer = TfidfTransformer()
    X_tfidf = tfidf_transformer.fit_transform(X_counts)
    naive_accurary = []
    rkf = RepeatedKFold(n_splits=10, n_repeats=2, random_state=42)


    model_name = []
    model_accuracy = []


    for model in models:
        for train, test in rkf.split(X):
            model.fit(X_tfidf[train], y[train])
            y_pred = model.predict(X_tfidf[test])
            model_name.append(model.__class__.__name__)
            model_accuracy.append(mt.accuracy_score(y[test], y_pred))
            
    disc_result = {'Model':  model_name, 'Accuracy': model_accuracy}      
    results = pd.DataFrame(disc_result, columns=["Model", "Accuracy"])   
    return results
```

```python
results = train(data)
```

```python
results
```

# Resultado de acurácia para cada modelo

```python
pd.DataFrame(results.groupby("Model").Accuracy.mean())
```

```python
def grafico(data):
    colors = ['royalblue', 'black', 'r', 'g']
    i = 0
    for model in models:
        acc = data[data['Model'] == model.__class__.__name__].Accuracy.values
        plt.plot(range(20), acc, 'b', label= model.__class__.__name__, color= colors[i])
        plt.legend()
        plt.title('Accuracy')    
        i = i+1

    plt.show()   
```

```python
grafico(results)
```

# Teste Hipótese Friedman

```python
def test(results):
    seed(1)
    data1 = results[results['Model'] == "LinearSVC"].Accuracy.values
    data2 = results[results['Model'] == "MultinomialNB"].Accuracy.values
    data3 = results[results['Model'] == "LogisticRegression"].Accuracy.values

    stat, p = friedmanchisquare(data1, data2, data3)
    print('Statistics=%.3f, p=%.10f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Falhou em Rejeitar H0')
    else:
        print('Rejeita H0')
```

```python
test(results)
```

# Repeintdo processo, entrentanto, agora considera-se a coluna IdSentenca

```python
data2 = pd.read_csv("Outputs/data2.csv")
y = data2.Sentimento.values
sentimentos = data2.groupby('Sentimento').Sentimento.count()
```
